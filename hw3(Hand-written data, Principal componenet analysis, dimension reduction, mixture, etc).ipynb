```python
import numpy as np

x=1
a = np.load('FilteredRecording.npy')
b = a>95
c = a<110.8
bool = b*c
indexOf = [i for i, x in enumerate(bool) if x]
l = []
for i in range(len(indexOf)):
    l.append(a[indexOf[i]-40:indexOf[i]+40])
```

```python
import matplotlib.pyplot as plt

plt.plot(l[0])
plt.show()
plt.plot(l[1000])
plt.show()
```

(b)

```python
l_2=[]
for i in range(len(l)):
    if np.argmax(l[i])>24 and np.argmax(l[i])<56:
        ind = np.argmax(l[i])
        l_2.append(l[i][ind-25:ind+25])
```

(c)

```python
from sklearn.cluster import KMeans
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")


kmeans = KMeans(n_clusters = 2)
kmeans.fit(l_2)
kmeans.predict(l_2)==1
centroids_2 = kmeans.cluster_centers_
plt.title("Centroids with two clusters")
plt.plot(centroids_2[0])
plt.plot(centroids_2[1])
plt.show()

p_2_0 = (kmeans.predict(l_2)==0).nonzero()[0]
for i in range(len(p_2_0)):
    plt.plot(l_2[p_2_0[i]], 'b')
plt.show()

p_2_1 = (kmeans.predict(l_2)==1).nonzero()[0]
for i in range(len(p_2_1)):
    plt.plot(l_2[p_2_1[i]], 'b')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(l_2)
centroids_3 = kmeans.cluster_centers_

plt.title("Centroids with three clusters")
plt.plot(centroids_3[0])
plt.plot(centroids_3[1])
plt.plot(centroids_3[2])
plt.show()

p_3_0 = (kmeans.predict(l_2)==0).nonzero()[0]
for i in range(len(p_3_0)):
    plt.plot(l_2[p_3_0[i]], 'b')
plt.show()

p_3_1 = (kmeans.predict(l_2)==1).nonzero()[0]
for i in range(len(p_3_1)):
    plt.plot(l_2[p_3_1[i]], 'b')
plt.show()

p_3_2 = (kmeans.predict(l_2)==2).nonzero()[0]
for i in range(len(p_3_2)):
    plt.plot(l_2[p_3_2[i]], 'b')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(l_2)
centroids_4 = kmeans.cluster_centers_
plt.title("Centroids with four clusters")
plt.plot(centroids_4[0])
plt.plot(centroids_4[1])
plt.plot(centroids_4[2])
plt.plot(centroids_4[3])
plt.show()

p_4_0 = (kmeans.predict(l_2)==0).nonzero()[0]
for i in range(len(p_4_0)):
    plt.plot(l_2[p_4_0[i]], 'b')
plt.show()

p_4_1 = (kmeans.predict(l_2)==1).nonzero()[0]
for i in range(len(p_4_1)):
    plt.plot(l_2[p_4_1[i]], 'b')
plt.show()

p_4_2 = (kmeans.predict(l_2)==2).nonzero()[0]
for i in range(len(p_4_2)):
    plt.plot(l_2[p_4_2[i]], 'b')
plt.show()

p_4_3 = (kmeans.predict(l_2)==3).nonzero()[0]
for i in range(len(p_4_3)):
    plt.plot(l_2[p_4_3[i]], 'b')
plt.show()
```

```python
from numpy import linalg as LA
def pca(vect, dim):
    X = np.cov(np.array(vect).T)
    a, b, c= np.linalg.svd(X)
    p = np.matmul(vect, c[:, 0:dim])
    return c, p, b
```

```python
c, p, b= pca(l_2,2)
plt.scatter(p[:,0], p[:,1], 0.3)
```
